```python
import numpy as np
from sympy import *
import pandas as pd

from ipywidgets import *
import ipywidgets as widgets
from IPython.display import clear_output
import ipysheet 
from ipysheet import sheet, cell, row, column, cell_range

import plotly.graph_objects as go
import plotly.graph_objects as go

pd.set_option('display.max_columns', None)
```

```python
def graficar(b,tipo):
    clear_output()
    display(tab)
    
    grafica = go.Figure()
    
    if tipo == 1: # por formula
        st = float(min.value)
        fn = float(max.value)
        stp = float(step.value)
        fun = form.value
        
        if tipo.value == 1:
            x = Symbol('x')
            y = eval(f"{fun}")
            xx = x
            yprime = y.diff(xx)

            x = np.arange(st,fn+stp,stp)
            f = lambdify(xx,yprime,'numpy')
            yprime = f(x)
            y = eval(f"{fun}")
            fun = fun.replace('**', '^').replace('*', '') #vel

            data = pd.DataFrame({
                "$t_o(s)$": x[:-1],
                "$t_f(s)$": x[1:],
                "$v_o$": y[:-1],
                "$v_f$": y[1:],
                "$\Delta t(s)$": x[1:]-x[:-1],
                "$\Delta v$":y[1:]-y[:-1] ,
                "$v_p$": (y[1:]+y[:-1])/2,
                "$\Delta x$": ((y[1:]+y[:-1])/2) *  (x[1:]-x[:-1]),
                "$x_o$":tipo_text.children[0].value,
                "$x_f$":(((y[1:]+y[:-1])/2) *  (x[1:]-x[:-1])) + tipo_text.children[0].value,
                "$|\Delta x|$": abs(((y[1:]+y[:-1])/2) *  (x[1:]-x[:-1])),
                "$a$":yprime[:-1]
            })
            for i in range(1, len(data) ):
                data["$x_o$"][i] = data["$x_f$"][i-1]
                data["$x_f$"][i] = data["$\Delta x$"][i] + data["$x_o$"][i]

            grafica.add_trace(go.Scatter(x=x, y=y, name=f"$V$",line_shape="linear"))
            grafica.add_trace(go.Scatter(x=x, y=data["$\Delta v$"], name=f"$\Delta v$",line_shape="linear"))
            grafica.add_trace(go.Scatter(x=x, y=data["$\Delta x$"], name=f"$\Delta x$",line_shape="linear"))
            grafica.add_trace(go.Scatter(x=x, y=data["$v_p$"], name=f"$v_p$",line_shape="linear"))
            grafica.add_trace(go.Scatter(x=x, y=data["$x_o$"], name="$X$",line_shape="linear"))
            grafica.add_trace(go.Scatter(x=x, y=data["$a$"], name="$a$",line_shape="linear"))
        else:
            x = Symbol('x')
            y = eval(f"{fun}")
            xx = x
            yprime = y.diff(xx)
            acy = yprime.diff(xx)

            x = np.arange(st,fn+stp,stp)
            f = lambdify(xx,yprime,'numpy')
            g = lambdify(xx,acy,'numpy')
            yprime = f(x)
            ac = g(x)
            y = eval(f"{fun}")
            fun = fun.replace('**', '^').replace('*', '') #pos

            data = pd.DataFrame({
                "$t_o(s)$": x[:-1],
                "$t_f(s)$": x[1:],
                "$x_o$":y[:-1],
                "$x_f$": y[1:],
                "$\Delta t(s)$": x[1:]-x[:-1],
                "$\Delta x$":y[1:]-y[:-1],
                "$v$":yprime[:-1],
                "$a$":ac[:-1]
            })

            grafica.add_trace(go.Scatter(x=x, y=y, name=f"$X$",line_shape="linear"))
            grafica.add_trace(go.Scatter(x=x, y=data["$\Delta x$"], name=f"$\Delta x$",line_shape="linear"))
            grafica.add_trace(go.Scatter(x=x, y=data["$v$"], name=f"$v$",line_shape="linear"))
            grafica.add_trace(go.Scatter(x=x, y=data["$a$"], name="$a$",line_shape="linear"))
    else: # arrbitraria
        print(x.value,y.value)
            
    #axs[0].table(cellText=data.values, colLabels=data.columns, loc='center',fontsize=90.0,colWidths=[.05]*10)
    #axs[0].scale(1, 4)
    #axs[0].axis("off")
    
   # print(data)

    grafica.show()
    
    tabla = go.Figure()
    tabla.add_trace(
        go.Table(
            header=dict(
                values=data.columns,
                font=dict(size=10),
                align="left"
            ),
            cells=dict(
                values=[data[k].tolist() for k in data.columns],
                align = "left")
        )
    )
    
    tabla.show()

def handle_check_change(change):
    tipo_text.children[1].disabled = not change.new
```

```python
min = widgets.FloatText(
    value=0,
    description='Minimo:',
    disabled=False,
    width='5%'
)
max = widgets.FloatText(
    value=10,
    description='Maximo:',
    disabled=False
)
step = widgets.FloatText(
    value=1,
    description='Pasos:',
    disabled=False
)
form = widgets.Text(
    value="x**2",
    description='Formula:',
    disabled=False
)
tipo_text = Box(
    [
        widgets.Checkbox(
            value=True,
            description='Posición inicial',
            disabled=False
        ),
        widgets.FloatText(
            value=0,
            #description='Posición inicial:',
            disabled=False
        )
    ]
)
tipo_text.children[0].observe(handle_check_change, names='value')

tipo = widgets.Dropdown(
    options=[('Velocidad - Tiempo',1), ('Posición - Tiempo',2)],
    value=1,
    width="100%",
    description='Grafico:',
    disabled=False,
)

run = widgets.Button(
        description='Graficar',
        disabled=False,
        button_style='success', # 'success', 'info', 'warning', 'danger' or ''
        tooltip='Graficar',
        icon='play' # (FontAwesome names without the `fa-` prefix),
)

items = widgets.GridBox([min,max,step,form,tipo_text,tipo,run], layout=widgets.Layout(grid_template_columns="repeat(2, 50%)"))

run.on_click(graficar,1)
```

```python
def preparar(b):
    graficar(2)
x = widgets.Text(
    value='0, 10, 20, 30',
    placeholder='0, 10, 20, 30',
    description='X:',
    disabled=False
)
y = widgets.Text(
    value='0, 10, 20, 30',
    placeholder='0, 10, 20, 30',
    description='Y:',
    disabled=False
)

run2 = widgets.Button(
        description='Graficar',
        disabled=False,
        button_style='success', # 'success', 'info', 'warning', 'danger' or ''
        tooltip='Graficar',
        icon='play' # (FontAwesome names without the `fa-` prefix),
)

arb = widgets.GridBox([x,y,tipo,run2], layout=widgets.Layout(grid_template_columns="repeat(2, 50%)"))
run2.on_click(graficar,2)
```

```python
tab = widgets.Tab(
    children=[
        items,
        arb],
    )
tab.set_title(0,"Formula")
tab.set_title(1,"Arbitrario")
display(tab)
```
